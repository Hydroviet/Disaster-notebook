```python
def plot_tif(tifFilePath, ax=None):
    import rasterio as rio
    import matplotlib.pyplot as plt
    with rio.open(tifFilePath) as src:
        ndvi = src.read(1)
        crs = src.crs
        trans = src.transform
        bounds = src.bounds
    left, top = bounds.left, bounds.top
    right, bottom = bounds.right, bounds.bottom
    if ax is None:
        fig = plt.figure()
        ax = fig.add_subplot(111)
    ax.imshow(ndvi, aspect=1.0, extent=[left, right, bottom, top])
    ax.set_title(tifFilePath)
    return ax
```

```python
def plot_tif_in_folder(folderPath):
    import os
    import rasterio
    import matplotlib.pyplot as plt
    
    tifs = None
    for _,_, files in os.walk(folderPath):
        tifs = files
    nFiles = len(tifs)
    fSize = nFiles*5
    fig = plt.figure(figsize=(fSize, fSize))
    axes = fig.subplots(nFiles, 1)
    for ax, tif in zip(axes, tifs):
        plot_tif(folderPath + '/' + tif, ax)
    
        
plot_tif_in_folder('/home/lamductan/Downloads/MODIS/2')
```
